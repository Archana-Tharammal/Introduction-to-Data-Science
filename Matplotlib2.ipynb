```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0, 0, 1,1])
ax.plot(x, y, 'b')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title');
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0, 0, 1, 1])
axes2 = fig.add_axes([0.2, 0.5, .2, .2])
```

```python
axes1.plot(x, y, 'r')
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes2.plot(y, x, 'r')
axes2.set_xlabel('x')
axes2.set_ylabel('y');
fig
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0, 0, 1, 1])
axes2 = fig.add_axes([0.2, 0.5, .4, .4])
```

```python

axes1.plot(x,z,'b')
axes1.set_xlabel('X')
axes1.set_ylabel('Z')

axes2.plot(x,y,'b')
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
axes2.set_title('zoom')
axes2.set_xlim([20, 22])
axes2.set_ylim([30, 50])

fig
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
```

```python
axes[0].plot(x, y, color="blue", lw=3, linestyle='--')
axes[1].plot(x, z, color="red", lw=3, linestyle='-')
fig
```

```python
fig, axes= plt.subplots(nrows=1, ncols=2,figsize=(12,2))
axes[0].plot(x, y, color="blue", lw=3)
axes[0].set_xlabel("x")
axes[0].set_ylabel("y")

axes[1].plot(x, z, color="red", lw=3, linestyle='--')
axes[1].set_xlabel("x")
axes[1].set_ylabel("z")
```
