```python
import numpy as np
import pandas as pd
```

```python
df=pd.DataFrame({'A':[1,2,np.nan],
                 'B':[5,np.nan,np.nan],
                  'C':[1,2,3]})
```

```python
df
```

```python
df.dropna()
```

```python
df.dropna(axis=1)
```

```python
df.dropna(thresh=2)
```

```python
df.fillna(value='FILL VALUE')
```

```python
df['A'].fillna(value=df['A'].mean())
```
