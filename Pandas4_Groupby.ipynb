```python
import pandas as pd
data=  {'Company':['GOOG','GOOG','MSFT','MSFT','FB','FB'],
       'Person':['Sam','Charlie','Amy','Vanessa','Carl','Sarah'],
       'Sales':[200,120,340,124,243,350]}
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python
df.groupby('Company')
```

```python
by_comp=df.groupby("Company")
```

```python
by_comp
```

```python
by_comp.mean()
```

```python
df.groupby("Company").mean()
```

```python
by_comp.std()
```

```python
by_comp.min()
```

```python
by_comp.max()
```

```python
by_comp.count()
```

```python
by_comp.describe()
```

```python
by_comp.describe().transpose()
```
